```python
import skrf
import matplotlib.pyplot as plt
from cal_kits import Keysight85056D, Keysight85058BP
```

```python
# 85056A Cal Kit

cal_ff_A = Keysight85056D.one_port_m(
    measured_f_short = skrf.Network('DIRECT/85056A_FSHORT_FieldFox_mate1_swp0.s1p'),
    measured_f_open  = skrf.Network('DIRECT/85056A_FOPEN_FieldFox_mate1_swp0.s1p'),
    measured_f_load  = skrf.Network('DIRECT/85056A_FBBLOAD_FieldFox_mate1_swp0.s1p')
)
cal_8510_A = Keysight85056D.one_port_m(
    measured_f_short = skrf.Network('DIRECT/85056A_FSHORT_8510C_mate1_swp0.s1p'),
    measured_f_open  = skrf.Network('DIRECT/85056A_FOPEN_8510C_mate1_swp0.s1p'),
    measured_f_load  = skrf.Network('DIRECT/85056A_FBBLOAD_8510C_mate1_swp0.s1p')
)



# 85058B Cal Kit (Polynomial, Broadband)

cal_ff_B = Keysight85058BP.one_port_m(
    measured_f_short1 = skrf.Network('DIRECT/85058B_FSHORT1_FieldFox_mate1_swp0.s1p'),
    measured_f_open   = skrf.Network('DIRECT/85058B_FOPEN_FieldFox_mate1_swp0.s1p'),
    measured_f_load   = skrf.Network('DIRECT/85058B_FLBLOAD_FieldFox_mate1_swp0.s1p'),
    use_coarse_definition=True
)
cal_8510_B = Keysight85058BP.one_port_m(
    measured_f_short1 = skrf.Network('DIRECT/85058B_FSHORT1_8510C_mate1_swp0.s1p'),
    measured_f_open   = skrf.Network('DIRECT/85058B_FOPEN_8510C_mate1_swp0.s1p'),
    measured_f_load   = skrf.Network('DIRECT/85058B_FLBLOAD_8510C_mate1_swp0.s1p'),
    use_coarse_definition=True
)
```

```python
cal_ff_A.apply_cal(skrf.Network('DIRECT/3654B_FSHORT_FieldFox_mate1_swp0.s1p', name='FieldFox, cal=85056D')).plot_s_db()
cal_ff_B.apply_cal(skrf.Network('DIRECT/3654B_FSHORT_FieldFox_mate1_swp0.s1p', name='FieldFox, cal=85058BP-BB')).plot_s_db()
cal_8510_A.apply_cal(skrf.Network('DIRECT/3654B_FSHORT_8510C_mate1_swp0.s1p',  name='8510C, cal=85056D')).plot_s_db()
cal_8510_B.apply_cal(skrf.Network('DIRECT/3654B_FSHORT_8510C_mate1_swp0.s1p',  name='8510C, cal=85058BP-BB')).plot_s_db()
plt.title('Anritsu FSHORT')
plt.show()
```
