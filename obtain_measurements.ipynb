```python
%%time
vna.set_frequency_sweep(45e6,50e9,10001)
for i in tqdm.tqdm([0,1,2]):
    fname = f'/tmp/SEA_3654B/3654B_FF_AnritsuPower_MA2475A_8510C_mate1_swp{i}.s1p'
    if os.path.exists(fname):
        print('!!! FILE EXISTS !!!')
        continue
    net = vna.get_snp_network(ports=(1,))
    #net.write_touchstone(filename=fname)
```

# FieldFox Fetch
